```python
def print_section_title(title):
    print(colored(title, 'blue', attrs=['bold', 'underline']))

def display_head_and_tail(dataframe, head=5):
    display(dataframe.head(head).style.set_caption("Head"))
    display(dataframe.tail(head).style.set_caption("Tail"))

def display_na(dataframe):
    na_df = dataframe.isnull().sum().reset_index()
    na_df.columns = ['Column', 'Number of NA']
    display(na_df.style.set_caption("Number of NA Values"))

def display_quantiles(dataframe):
    quantiles_df = dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T
    display(quantiles_df.style.format("{:.2f}").set_caption("Quantiles"))

def check_df(dataframe, head=5):
    print_section_title('Shape')
    print(dataframe.shape)
    print_section_title('Types')
    print(dataframe.dtypes.to_frame('Data Type').style.set_caption("Data Types"))
    print_section_title('Head & Tail')
    display_head_and_tail(dataframe, head)
    print_section_title('NA Values')
    display_na(dataframe)
    print_section_title('Quantiles')
    display_quantiles(dataframe)

check_df(df_13)
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Global Solar Energy Capacity Over Years</div></b>

```python
global_trend = df_13.groupby('Year')['Solar Capacity'].sum().reset_index()

frames = [go.Frame(
    data=[go.Scatter(x=global_trend['Year'][:k+1], y=global_trend['Solar Capacity'][:k+1], mode='lines+markers')],
    layout=go.Layout(title_text=f'Global Solar Energy Capacity Until {year}')
) for k, year in enumerate(global_trend['Year'])]

initial_frame = go.Frame(
    data=[go.Scatter(x=[global_trend['Year'].iloc[0]], y=[global_trend['Solar Capacity'].iloc[0]], mode='markers')],
    layout=go.Layout(title_text='Global Solar Energy Capacity (Initial Year)')
)
frames.insert(0, initial_frame)

fig = go.Figure(
    data=[go.Scatter(x=global_trend['Year'], y=global_trend['Solar Capacity'], mode='lines+markers')],
    layout=go.Layout(
        xaxis=dict(range=[global_trend['Year'].min(), global_trend['Year'].max()], autorange=False),
        yaxis=dict(range=[global_trend['Solar Capacity'].min(), global_trend['Solar Capacity'].max()], autorange=False),
        title="Global Solar Energy Capacity Over Years",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args=[None])])]
    ),
    frames=frames
)

fig.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Solar Energy Capacity Ranking in Europe in 2021</div></b>

```python
latest_year_data = df_13[df_13['Year'] == df_13['Year'].max()]

european_countries = ['Germany', 'Spain', 'United Kingdom', 'France', 'Italy', 'Sweden', 'Poland', 'Netherlands', 'Turkey']

europe_latest_year = latest_year_data[latest_year_data['Entity'].isin(european_countries)].sort_values(by='Solar Capacity', ascending=False)

fig = px.bar(europe_latest_year, y='Entity', x='Solar Capacity', 
             title=f'Solar Energy Capacity Ranking in Europe in {df_13["Year"].max()}',
             labels={'Entity': 'Country', 'Solar Capacity': 'Solar Capacity (GW)'},
             color='Solar Capacity',
             color_continuous_scale='Viridis',
             orientation='h')

fig.update_layout(
    xaxis_title='Solar Capacity (GW)',
    yaxis_title='Country',
    coloraxis_colorbar=dict(title='Solar Capacity (GW)'),
    plot_bgcolor='rgba(0,0,0,0)',
    template='plotly_white'
)

fig.show()
```

```python
yearly_increase = df_13.groupby(['Year', 'Entity'])['Solar Capacity'].sum().unstack().fillna(0).diff().fillna(0)

top_entities = yearly_increase.sum(axis=0).sort_values(ascending=False).head(5).index

filtered_yearly_increase = yearly_increase[top_entities]

filtered_yearly_increase.plot(kind='bar', figsize=(15, 8), width=0.7)
plt.title('Annual Increase in Solar Power Capacity by Region')
plt.xlabel('Year')
plt.ylabel('Increase in Solar Capacity (GW)')
plt.xticks(rotation=90)
plt.legend(title='Region')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Total Solar Power Production of Specified Countries</div></b>

```python
specified_countries = ['Austria', 'Belgium', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Italy', 
                      'Netherlands', 'Norway', 'Poland', 'Portugal', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom']

specified_countries_solar_data = df_13[df_13['Entity'].isin(specified_countries)]

total_solar_production_specified_countries = specified_countries_solar_data.groupby('Entity')['Solar Capacity'].sum().sort_values(ascending=False)

plt.figure(figsize=(20, 8))
bars = plt.bar(total_solar_production_specified_countries.index, total_solar_production_specified_countries.values, color='grey')
plt.title('Total Solar Power Production of Specified Countries', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Electricity from Solar (TWh)', fontsize=12, fontweight='bold')
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

for bar in bars:
    if bar.get_x() + bar.get_width() / 2 == total_solar_production_specified_countries.index.get_loc("Turkey"):
        bar.set_color('blue')
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

<a id='YearlyRegionalSharesofSolarPVCapacity'></a>
# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Yearly Regional Shares of Solar PV Capacity</div></b>

```python
regional_countries = ['Turkey', 'Germany', 'United States', 'China', 'India', 'Brazil']

regional_data = df_13[df_13['Entity'].isin(regional_countries)]

yearly_totals = regional_data.groupby('Year')['Solar Capacity'].sum().reset_index()

regional_data = regional_data.merge(yearly_totals, on='Year', suffixes=('', '_Total'))
regional_data['Percentage Share'] = (regional_data['Solar Capacity'] / regional_data['Solar Capacity_Total']) * 100

pivot_share = regional_data.pivot(index="Year", columns="Entity", values="Percentage Share")

pivot_share.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='viridis')
plt.title('Yearly Regional Shares of Solar PV Capacity', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage Share of Solar Capacity', fontsize=12)
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

The color differentiation for each country within the stacks allows for easy comparison across years, highlighting shifts in regional dominance and growth in solar capacity. This visualization underscores the dynamic nature of the global solar energy market, reflecting policy impacts, technological advancements, and investment trends in renewable energy across these pivotal nations.
