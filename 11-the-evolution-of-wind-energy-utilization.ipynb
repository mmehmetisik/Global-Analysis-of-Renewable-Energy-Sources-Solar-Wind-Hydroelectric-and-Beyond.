```python
def print_section_title(title):
    print(colored(title, 'blue', attrs=['bold', 'underline']))

def display_head_and_tail(dataframe, head=5):
    display(dataframe.head(head).style.set_caption("Head"))
    display(dataframe.tail(head).style.set_caption("Tail"))

def display_na(dataframe):
    na_df = dataframe.isnull().sum().reset_index()
    na_df.columns = ['Column', 'Number of NA']
    display(na_df.style.set_caption("Number of NA Values"))

def display_quantiles(dataframe):
    quantiles_df = dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T
    display(quantiles_df.style.format("{:.2f}").set_caption("Quantiles"))

def check_df(dataframe, head=5):
    print_section_title('Shape')
    print(dataframe.shape)
    print_section_title('Types')
    print(dataframe.dtypes.to_frame('Data Type').style.set_caption("Data Types"))
    print_section_title('Head & Tail')
    display_head_and_tail(dataframe, head)
    print_section_title('NA Values')
    display_na(dataframe)
    print_section_title('Quantiles')
    display_quantiles(dataframe)

check_df(df_11)
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Wind Power Usage By Continent</div></b>

```python
continent_dict = {
    'Africa': ['Algeria', 'Egypt', 'South Africa', 'Morocco', 'Nigeria'],
    'Asia': ['China', 'India', 'Japan', 'South Korea', 'Saudi Arabia'],
    'Europe': ['Germany', 'United Kingdom', 'France', 'Italy', 'Spain'],
    'North America': ['United States', 'Canada', 'Mexico'],
    'South America': ['Brazil', 'Argentina', 'Chile', 'Colombia'],
    'Oceania': ['Australia', 'New Zealand']
}

def classify_continent(country):
    for continent, countries in continent_dict.items():
        if country in countries:
            return continent
    return 'Other'

df_11['Continent'] = df_11['Entity'].apply(classify_continent)

continent_wind_usage = df_11.groupby(['Continent', 'Year'])['Wind (% electricity)'].mean().reset_index()

fig = px.line(continent_wind_usage, x='Year', y='Wind (% electricity)', color='Continent', 
              title='Wind Power Usage by Continent',
              labels={'Continent': 'Continent', 'Year': 'Year', 'Wind (% electricity)': 'Average Wind Power Usage (%)'},
              markers=True, line_shape='spline')

fig.update_layout(legend_title_text='Continent',
                  xaxis_title='Year',
                  yaxis_title='Average Wind Power Usage (%)',
                  xaxis=dict(tickangle=45),
                  template='plotly_white')

fig.update_traces(mode='markers+lines', marker=dict(size=8, line=dict(width=1)), line=dict(width=2.5))

fig.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Total Wind Power Production Share of Specified Countries</div></b>

```python
specified_countries = ["China", "Canada", "United States", "Brazil", "Russia", "Norway",
                       "Japan", "India", "Sweden", "France", "Italy", "Venezuela",
                       "Austria", "Switzerland", "Spain", "Turkey", "Colombia", "Mexico", "New Zealand", "Paraguay"]

specified_countries_data = df_11[df_11['Entity'].isin(specified_countries)]

total_wind_production_specified_countries = specified_countries_data.groupby('Entity')['Wind (% electricity)'].sum().sort_values(ascending=False)

plt.figure(figsize=(20, 8))
bars = plt.bar(total_wind_production_specified_countries.index, total_wind_production_specified_countries.values, color='skyblue')
plt.title('Total Wind Power Production Share of Specified Countries', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Wind (% electricity)', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

for bar in bars:
    if bar.get_x() + bar.get_width() / 2 == total_wind_production_specified_countries.index.get_loc("Turkey"):
        bar.set_color('red')
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

<a id='WindPowerProductionPercentagesOverDifferentYears(PolarAreaChart)'></a>
# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Wind Power Production Percentages Over Different Years (Polar Area Chart)</div></b>

```python
df_11_polar_area = df_11.groupby('Year')['Wind (% electricity)'].mean().reset_index()

df_11_polar_area['Year'] = df_11_polar_area['Year'].astype('category')

num_vars_polar = df_11_polar_area['Year'].nunique()

angles_polar = np.linspace(0, 2 * np.pi, num_vars_polar, endpoint=False).tolist()

stats_polar = df_11_polar_area['Wind (% electricity)'].tolist()

stats_polar += stats_polar[:1]
angles_polar += angles_polar[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
ax.fill(angles_polar, stats_polar, color='skyblue', alpha=0.25)
ax.plot(angles_polar, stats_polar, color='blue', linewidth=2)

ax.set_xticks(angles_polar[:-1])
ax.set_xticklabels(df_11_polar_area['Year'])

plt.title('Wind Power Production Percentages Over Different Years (Polar Area Chart)', size=16, color='black', y=1.1)

plt.show()
```

The circular format, with its radiating lines and filled color, emphasizes the cyclical nature and fluctuating trends of wind energy contribution to the electricity grid. This visualization method effectively captures the ebb and flow of wind power usage, with each year positioned along the chart's circumference, allowing for an intuitive comparison of data points in a continuous, looped timeline. The choice of sky blue evokes a sense of harmony with the theme of wind energy, while the chart as a whole offers a unique perspective on the progression and variability of wind power over time.
