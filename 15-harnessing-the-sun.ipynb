<a id='SolarPowerProductioninTurkey2021'></a>
# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Solar Power Production in Turkey 2021</div></b>

```python
turkey_data = df_15[df_15['Entity'] == 'Turkey']

frames = [go.Frame(
    data=[go.Scatter(x=turkey_data['Year'][:k+1], y=turkey_data['Solar (% electricity)'][:k+1], mode='lines+markers')],
    layout=go.Layout(title_text=f'Solar Power Production in Turkey ({year})')
) for k, year in enumerate(turkey_data['Year'])]

initial_frame = go.Frame(
    data=[go.Scatter(x=[turkey_data['Year'].iloc[0]], y=[turkey_data['Solar (% electricity)'].iloc[0]], mode='markers')],
    layout=go.Layout(title_text='Solar Power Production in Turkey (Initial Year)')
)
frames.insert(0, initial_frame)

fig = go.Figure(
    data=[go.Scatter(x=turkey_data['Year'], y=turkey_data['Solar (% electricity)'], mode='lines+markers')],
    layout=go.Layout(
        xaxis=dict(range=[turkey_data['Year'].min(), turkey_data['Year'].max()], autorange=False),
        yaxis=dict(range=[0, turkey_data['Solar (% electricity)'].max() * 1.1], autorange=False),  
        title="Solar Power Production in Turkey",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args=[None, {"frame": {"duration": 500, "redraw": True}, "fromcurrent": True, "transition": {"duration": 300, "easing": "quadratic-in-out"}}])]
        )]
    ),
    frames=frames
)

fig.show()
```

<a id='TotalSolarPowerProductionShareofSpecifiedCountries'></a>
# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Total Solar Power Production Share of Specified Countries</div></b>

```python
specified_countries = ["China", "Canada", "United States", "Brazil", "Russia", "Norway",
                       "Japan", "India", "Sweden", "France", "Italy", "Venezuela",
                       "Austria", "Switzerland", "Spain", "Turkey", "Colombia", "Mexico", "New Zealand", "Paraguay"]

specified_countries_data = df_15[df_15['Entity'].isin(specified_countries)]

total_solar_production_specified_countries = specified_countries_data.groupby('Entity')['Solar (% electricity)'].sum().sort_values(ascending=False)

plt.figure(figsize=(20, 8))
bars = plt.bar(total_solar_production_specified_countries.index, total_solar_production_specified_countries.values, color='skyblue')
plt.title('Total Solar Power Production Share of Specified Countries', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Solar (% electricity)', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

for bar in bars:
    if bar.get_x() + bar.get_width() / 2 == total_solar_production_specified_countries.index.get_loc("Turkey"):
        bar.set_color('red')
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Solar Power Production Percentages Over Different Years (Polar Area Chart)</div></b>

```python
df_15_polar_area = df_15.groupby('Year')['Solar (% electricity)'].mean().reset_index()

df_15_polar_area['Year'] = df_15_polar_area['Year'].astype('category')

num_vars_polar = df_15_polar_area['Year'].nunique()

angles_polar = np.linspace(0, 2 * np.pi, num_vars_polar, endpoint=False).tolist()

stats_polar = df_15_polar_area['Solar (% electricity)'].tolist()

stats_polar += stats_polar[:1]
angles_polar += angles_polar[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
ax.fill(angles_polar, stats_polar, color='skyblue', alpha=0.25)
ax.plot(angles_polar, stats_polar, color='blue', linewidth=2)

ax.set_xticks(angles_polar[:-1])
ax.set_xticklabels(df_15_polar_area['Year'])

plt.title('Solar Power Production Percentages Over Different Years (Polar Area Chart)', size=16, color='black', y=1.1)

plt.show()
```

The sky-blue fill, coupled with a solid blue outline, visually represents the fluctuations in solar energy production, with each radial segment corresponding to a different year. This format offers a unique perspective on the data, allowing for an intuitive comparison of yearly averages and the identification of trends over time. The chart effectively captures the cyclical nature of solar power growth, highlighting periods of significant increase or stability in solar energy production.
