```python
def print_section_title(title):
    print(colored(title, 'blue', attrs=['bold', 'underline']))

def display_head_and_tail(dataframe, head=5):
    display(dataframe.head(head).style.set_caption("Head"))
    display(dataframe.tail(head).style.set_caption("Tail"))

def display_na(dataframe):
    na_df = dataframe.isnull().sum().reset_index()
    na_df.columns = ['Column', 'Number of NA']
    display(na_df.style.set_caption("Number of NA Values"))

def display_quantiles(dataframe):
    quantiles_df = dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T
    display(quantiles_df.style.format("{:.2f}").set_caption("Quantiles"))

def check_df(dataframe, head=5):
    print_section_title('Shape')
    print(dataframe.shape)
    print_section_title('Types')
    print(dataframe.dtypes.to_frame('Data Type').style.set_caption("Data Types"))
    print_section_title('Head & Tail')
    display_head_and_tail(dataframe, head)
    print_section_title('NA Values')
    display_na(dataframe)
    print_section_title('Quantiles')
    display_quantiles(dataframe)

check_df(df_12)
```

```python
turkey_solar_data = df_12[df_12['Entity'] == 'Turkey']

frames = [go.Frame(
    data=[go.Scatter(x=turkey_solar_data['Year'][:k+1], y=turkey_solar_data['Electricity from solar (TWh)'][:k+1], mode='lines+markers')],
    layout=go.Layout(title_text=f'Solar Power Production in Turkey ({year})')
) for k, year in enumerate(turkey_solar_data['Year'])]

initial_frame = go.Frame(
    data=[go.Scatter(x=[turkey_solar_data['Year'].iloc[0]], y=[turkey_solar_data['Electricity from solar (TWh)'].iloc[0]], mode='markers')],
    layout=go.Layout(title_text='Solar Power Production in Turkey (First Recorded Year)')
)
frames.insert(0, initial_frame)

fig = go.Figure(
    data=[go.Scatter(x=turkey_solar_data['Year'], y=turkey_solar_data['Electricity from solar (TWh)'], mode='lines+markers')],
    layout=go.Layout(
        xaxis=dict(range=[turkey_solar_data['Year'].min(), turkey_solar_data['Year'].max()], autorange=False),
        yaxis=dict(range=[0, turkey_solar_data['Electricity from solar (TWh)'].max() + turkey_solar_data['Electricity from solar (TWh)'].max()*0.1], autorange=False),
        title="Solar Power Production in Turkey",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args=[None])])]
    ),
    frames=frames
)

fig.show()
```

<a id='TotalSolarPowerProductionofSpecifiedCountries'></a>
# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Total Solar Power Production of Specified Countries</div></b>

```python
specified_countries = ["China", "United States", "India", "Japan", "Germany", "Italy", "United Kingdom", "France", "Australia", "Spain", "Turkey", "Brazil", "Canada", "South Korea", "Mexico", "Netherlands", "Switzerland", "Sweden", "Belgium", "Poland"]

specified_countries_data = df_12[df_12['Entity'].isin(specified_countries)]

total_solar_production_specified_countries = specified_countries_data.groupby('Entity')['Electricity from solar (TWh)'].sum().sort_values(ascending=False)

plt.figure(figsize=(20, 8))
bars = plt.bar(total_solar_production_specified_countries.index, total_solar_production_specified_countries.values, color='skyblue')
plt.title('Total Solar Power Production of Specified Countries', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Electricity from Solar (TWh)', fontsize=12, fontweight='bold')
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

for bar in bars:
    if bar.get_x() + bar.get_width() / 2 == total_solar_production_specified_countries.index.get_loc("Turkey"):
        bar.set_color('red')
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Solar Electricity Generation in Selected Countries Over Decades</div></b>

```python
selected_countries = ['Germany', 'France', 'Italy', 'Spain', 'United Kingdom', 
                      'Greece', 'Bulgaria', 'Georgia', 'Iran', 'Iraq',        
                      'Turkey', 'United States']
def assign_decade(year):
    for decade in decades:
        if year >= decade and year < decade + 10:
            return decade
    return "2020s" 

decades = list(range(1960, 2020, 10))

production_df_selected = df_12[df_12['Entity'].isin(selected_countries)].copy()

production_df_selected.fillna(0, inplace=True)

production_df_selected['Decade'] = production_df_selected['Year'].apply(assign_decade)

grouped_production = production_df_selected.groupby(['Entity', 'Decade'])['Electricity from solar (TWh)'].sum().unstack(fill_value=0)

grouped_production.plot(kind='bar', figsize=(15, 10))
plt.title("Solar Electricity Generation in Selected Countries Over Decades", fontweight='bold')
plt.xlabel("Countries", fontweight='bold')
plt.ylabel("Total Electricity from Solar (TWh)", fontweight='bold')
plt.legend(title="Decade")
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Solar Power Production Over Different Years (Polar Area Chart)</div></b>

```python
df_12_polar_area = df_12.groupby('Year')['Electricity from solar (TWh)'].mean().reset_index()

df_12_polar_area['Year'] = df_12_polar_area['Year'].astype('category')

num_vars_polar = df_12_polar_area['Year'].nunique()

angles_polar = np.linspace(0, 2 * np.pi, num_vars_polar, endpoint=False).tolist()

stats_polar = df_12_polar_area['Electricity from solar (TWh)'].tolist()

stats_polar += stats_polar[:1]
angles_polar += angles_polar[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
ax.fill(angles_polar, stats_polar, color='skyblue', alpha=0.25)
ax.plot(angles_polar, stats_polar, color='blue', linewidth=2)

ax.set_xticks(angles_polar[:-1])
ax.set_xticklabels(df_12_polar_area['Year'])

plt.title('Solar Power Production Over Different Years (Polar Area Chart)', size=16, color='black', y=1.1)

plt.show()
```
