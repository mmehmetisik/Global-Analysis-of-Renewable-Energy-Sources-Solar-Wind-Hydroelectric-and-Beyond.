# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Exploratory Data Analysis</div></b>

```python
def print_section_title(title):
    print(colored(title, 'blue', attrs=['bold', 'underline']))

def display_head_and_tail(dataframe, head=5):
    display(dataframe.head(head).style.set_caption("Head"))
    display(dataframe.tail(head).style.set_caption("Tail"))

def display_na(dataframe):
    na_df = dataframe.isnull().sum().reset_index()
    na_df.columns = ['Column', 'Number of NA']
    display(na_df.style.set_caption("Number of NA Values"))

def display_quantiles(dataframe):
    quantiles_df = dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T
    display(quantiles_df.style.format("{:.2f}").set_caption("Quantiles"))

def check_df(dataframe, head=5):
    print_section_title('Shape')
    print(dataframe.shape)
    print_section_title('Types')
    print(dataframe.dtypes.to_frame('Data Type').style.set_caption("Data Types"))
    print_section_title('Head & Tail')
    display_head_and_tail(dataframe, head)
    print_section_title('NA Values')
    display_na(dataframe)
    print_section_title('Quantiles')
    display_quantiles(dataframe)

check_df(df_16)
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Annual Volatility in Biofuel Production</div></b>

```python
yearly_volatility = df_16.groupby('Year')['Biofuels Production - TWh - Total'].std().reset_index()

fig = go.Figure(data=[
    go.Bar(
        x=yearly_volatility['Year'],
        y=yearly_volatility['Biofuels Production - TWh - Total'],
        marker_color='orange'
    )
])

fig.update_layout(
    title='Annual Volatility in Biofuel Production',
    xaxis=dict(title='Year', tickmode='linear'),
    yaxis=dict(title='Standard Deviation of Biofuels Production (TWh)'),
    showlegend=False
)

fig.show()
```

<a id='TotalBiofuelProductionofSpecifiedCountries'></a>
# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Total Biofuel Production of Specified Countries</div></b>

```python
specified_countries = ["China", "United States", "India", "Japan", "Germany", "Italy", "United Kingdom", "France", "Australia", "Spain", "United States", "Brazil", "Canada", "South Korea", "Mexico", "Netherlands", "Switzerland", "Sweden", "Belgium", "Poland"]

specified_countries_data = df_16[df_16['Entity'].isin(specified_countries)]

total_biofuel_production_specified_countries = specified_countries_data.groupby('Entity')['Biofuels Production - TWh - Total'].sum().sort_values(ascending=False)

plt.figure(figsize=(20, 8))
bars = plt.bar(total_biofuel_production_specified_countries.index, total_biofuel_production_specified_countries.values, color='skyblue')
plt.title('Total Biofuel Production of Specified Countries', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Biofuels Production (TWh)', fontsize=12, fontweight='bold')
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

for bar in bars:
    if bar.get_x() + bar.get_width() / 2 == total_biofuel_production_specified_countries.index.get_loc("United States"):
        bar.set_color('red')
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Biofuel Production Over Different Years (Polar Area Chart)</div></b>

```python
df_16_polar_area = df_16.groupby('Year')['Biofuels Production - TWh - Total'].mean().reset_index()

df_16_polar_area['Year'] = df_16_polar_area['Year'].astype('category')

num_vars_polar = df_16_polar_area['Year'].nunique()

angles_polar = np.linspace(0, 2 * np.pi, num_vars_polar, endpoint=False).tolist()

stats_polar = df_16_polar_area['Biofuels Production - TWh - Total'].tolist()

stats_polar += stats_polar[:1]
angles_polar += angles_polar[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
ax.fill(angles_polar, stats_polar, color='skyblue', alpha=0.25)
ax.plot(angles_polar, stats_polar, color='blue', linewidth=2) 

ax.set_xticks(angles_polar[:-1])
ax.set_xticklabels(df_16_polar_area['Year'])

plt.title('Biofuel Production Over Different Years (Polar Area Chart)', size=16, color='black', y=1.1)

plt.show()
```

The shades of blue and the radial spread of the data points encapsulate the fluctuations and trends in biofuel production. This visualization method offers a unique perspective, highlighting the cyclical nature and potential seasonality in biofuel output, while also emphasizing years of significant production through the broader spans of the chart.
