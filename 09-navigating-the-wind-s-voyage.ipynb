# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Exploratory Data Analysis</div></b>

```python
def print_section_title(title):
    print(colored(title, 'blue', attrs=['bold', 'underline']))

def display_head_and_tail(dataframe, head=5):
    display(dataframe.head(head).style.set_caption("Head"))
    display(dataframe.tail(head).style.set_caption("Tail"))

def display_na(dataframe):
    na_df = dataframe.isnull().sum().reset_index()
    na_df.columns = ['Column', 'Number of NA']
    display(na_df.style.set_caption("Number of NA Values"))

def display_quantiles(dataframe):
    quantiles_df = dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T
    display(quantiles_df.style.format("{:.2f}").set_caption("Quantiles"))

def check_df(dataframe, head=5):
    print_section_title('Shape')
    print(dataframe.shape)
    print_section_title('Types')
    print(dataframe.dtypes.to_frame('Data Type').style.set_caption("Data Types"))
    print_section_title('Head & Tail')
    display_head_and_tail(dataframe, head)
    print_section_title('NA Values')
    display_na(dataframe)
    print_section_title('Quantiles')
    display_quantiles(dataframe)

check_df(df_9)
```

<a id='GlobalWindEnergyCapacityOverYears'></a>
# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Global Wind Energy Capacity Over Years</div></b>

```python
global_trend = df_9.groupby('Year')['Wind Capacity'].sum().reset_index()

frames = [go.Frame(
    data=[go.Scatter(x=global_trend['Year'][:k+1], y=global_trend['Wind Capacity'][:k+1], mode='lines+markers')],
    layout=go.Layout(title_text=f'Global Wind Energy Capacity Until {year}')
) for k, year in enumerate(global_trend['Year'])]

initial_frame = go.Frame(
    data=[go.Scatter(x=[global_trend['Year'].iloc[0]], y=[global_trend['Wind Capacity'].iloc[0]], mode='markers')],
    layout=go.Layout(title_text='Global Wind Energy Capacity (Initial Year)')
)
frames.insert(0, initial_frame)

fig = go.Figure(
    data=[go.Scatter(x=global_trend['Year'], y=global_trend['Wind Capacity'], mode='lines+markers')],
    layout=go.Layout(
        xaxis=dict(range=[global_trend['Year'].min(), global_trend['Year'].max()], autorange=False),
        yaxis=dict(range=[global_trend['Wind Capacity'].min(), global_trend['Wind Capacity'].max()], autorange=False),
        title="Global Wind Energy Capacity Over Years",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args=[None])])]
    ),
    frames=frames
)

fig.show()
```

```python
yearly_increase = df_9.groupby(['Year', 'Entity'])['Wind Capacity'].sum().unstack().fillna(0).diff().fillna(0)

top_entities = yearly_increase.sum(axis=0).sort_values(ascending=False).head(5).index

filtered_yearly_increase = yearly_increase[top_entities]

filtered_yearly_increase.plot(kind='bar', figsize=(15, 8), width=0.7)
plt.title('Annual Increase in Wind Power Capacity by Region')
plt.xlabel('Year')
plt.ylabel('Increase in Wind Capacity (GW)')
plt.xticks(rotation=90)
plt.legend(title='Region')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Total Wind Power Production of Specified Countries</div></b>

```python
specified_countries = ['Austria', 'Belgium', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Italy', 
                      'Netherlands', 'Norway', 'Poland', 'Portugal', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom']

specified_countries_wind_data = df_9[df_9['Entity'].isin(specified_countries)]

total_wind_production_specified_countries = specified_countries_wind_data.groupby('Entity')['Wind Capacity'].sum().sort_values(ascending=False)

plt.figure(figsize=(20, 8))
bars = plt.bar(total_wind_production_specified_countries.index, total_wind_production_specified_countries.values, color='grey')
plt.title('Total Wind Power Production of Specified Countries', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Electricity from Wind (TWh)', fontsize=12, fontweight='bold')
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

for bar in bars:
    if bar.get_x() + bar.get_width() / 2 == total_wind_production_specified_countries.index.get_loc("Turkey"):
        bar.set_color('blue')
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
