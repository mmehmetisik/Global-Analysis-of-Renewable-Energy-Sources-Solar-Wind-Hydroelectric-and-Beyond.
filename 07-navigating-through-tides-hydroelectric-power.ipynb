# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Exploratory Data Analysis</div></b>

```python
def print_section_title(title):
    print(colored(title, 'blue', attrs=['bold', 'underline']))

def display_head_and_tail(dataframe, head=5):
    display(dataframe.head(head).style.set_caption("Head"))
    display(dataframe.tail(head).style.set_caption("Tail"))

def display_na(dataframe):
    na_df = dataframe.isnull().sum().reset_index()
    na_df.columns = ['Column', 'Number of NA']
    display(na_df.style.set_caption("Number of NA Values"))

def display_quantiles(dataframe):
    quantiles_df = dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T
    display(quantiles_df.style.format("{:.2f}").set_caption("Quantiles"))

def check_df(dataframe, head=5):
    print_section_title('Shape')
    print(dataframe.shape)
    print_section_title('Types')
    print(dataframe.dtypes.to_frame('Data Type').style.set_caption("Data Types"))
    print_section_title('Head & Tail')
    display_head_and_tail(dataframe, head)
    print_section_title('NA Values')
    display_na(dataframe)
    print_section_title('Quantiles')
    display_quantiles(dataframe)

check_df(df_7)
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Hydroelectric Power Production in Turkey</div></b>

```python
turkey_data = df_7[df_7['Entity'] == 'Turkey']

frames = [go.Frame(
    data=[go.Scatter(x=turkey_data['Year'][:k+1], y=turkey_data['Hydro (% electricity)'][:k+1], mode='lines+markers')],
    layout=go.Layout(title_text=f'Hydroelectric Power Production in Turkey ({year})')
) for k, year in enumerate(turkey_data['Year'])]

initial_frame = go.Frame(
    data=[go.Scatter(x=[turkey_data['Year'].iloc[0]], y=[turkey_data['Hydro (% electricity)'].iloc[0]], mode='markers')],
    layout=go.Layout(title_text='Hydroelectric Power Production in Turkey (1965)')
)
frames.insert(0, initial_frame)

fig = go.Figure(
    data=[go.Scatter(x=turkey_data['Year'], y=turkey_data['Hydro (% electricity)'], mode='lines+markers')],
    layout=go.Layout(
        xaxis=dict(range=[turkey_data['Year'].min(), turkey_data['Year'].max()], autorange=False),
        yaxis=dict(range=[turkey_data['Hydro (% electricity)'].min(), turkey_data['Hydro (% electricity)'].max()], autorange=False),
        title="Hydroelectric Power Production in Turkey",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args=[None, {"frame": {"duration": 500, "redraw": True}, "fromcurrent": True, "transition": {"duration": 300, "easing": "quadratic-in-out"}}])]
        )]
    ),
    frames=frames
)
fig
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Total Hydroelectric Power Production of Specified Countries</div></b>

```python
specified_countries = ["China", "Canada", "United States", "Brazil", "Russia", "Norway",
                       "Japan", "India", "Sweden", "France", "Italy", "Venezuela",
                       "Austria", "Switzerland", "Spain", "Turkey", "Colombia", "Mexico", "New Zealand", "Paraguay"]

specified_countries_data = df_7[df_7['Entity'].isin(specified_countries)]

total_production_specified_countries = specified_countries_data.groupby('Entity')['Hydro (% electricity)'].sum().sort_values(ascending=False)

plt.figure(figsize=(20, 8))
bars = plt.bar(total_production_specified_countries.index, total_production_specified_countries.values, color='grey')
plt.title('Total Hydroelectric Power Production of Specified Countries', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Hydro (% electricity)', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

for bar in bars:
    if bar.get_x() + bar.get_width() / 2 == total_production_specified_countries.index.get_loc("Turkey"):
        bar.set_color('blue')
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
decades = list(range(1960, 2020, 10))

def assign_decade(year):
    for decade in decades:
        if year >= decade and year < decade + 10:
            return decade
    return "2020s"

df_7['Decade'] = df_7['Year'].apply(assign_decade)

total_production_by_decade = df_7.groupby('Decade')['Hydro (% electricity)'].sum().reset_index()

fig = px.bar(total_production_by_decade, x='Decade', y='Hydro (% electricity)',
             hover_data=['Decade'], color='Hydro (% electricity)',
             labels={'Decade': 'Decade', 'Hydro (% electricity)': 'Total Electricity from Hydro (%)'}, height=400)

fig.update_layout(
    title='Global Hydroelectric Power Production by Decade',
    xaxis_title='Decade',
    yaxis_title='Total Electricity from Hydro (%)'
)

fig
```

```python
periods = [(2000, 2009), (2010, 2022)]

production_by_period = {}

for start, end in periods:
    mask = (df_7['Year'] >= start) & (df_7['Year'] <= end)
    total_production = df_7[mask].groupby('Entity')['Hydro (% electricity)'].sum()
    production_by_period[f"{start}-{end}"] = total_production

production_df = pd.DataFrame(production_by_period)

production_df.fillna(0, inplace=True)

top_countries = production_df.sum(axis=1).sort_values(ascending=False).head(20).index
production_df_top = production_df.loc[top_countries]

production_df_top.plot(kind='bar', figsize=(20, 10))
plt.title("Hydroelectricity Generation of Categorized Countries in Ten-Year Periods", fontweight='bold')
plt.xlabel("Categorized Countries", fontweight='bold')
plt.ylabel("Total Electricity from Hydro (%)", fontweight='bold')
plt.legend(title="Period")
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Hydroelectric Energy Production Percentages Over Different Years (Polar Area Chart)</div></b>

```python
df_7_polar_area = df_7.groupby('Year')['Hydro (% electricity)'].mean().reset_index()

df_7_polar_area['Year'] = df_7_polar_area['Year'].astype('category')

num_vars_polar = df_7_polar_area['Year'].nunique()

angles_polar = np.linspace(0, 2 * np.pi, num_vars_polar, endpoint=False).tolist()
stats_polar = df_7_polar_area['Hydro (% electricity)'].tolist()

stats_polar += stats_polar[:1]
angles_polar += angles_polar[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
ax.fill(angles_polar, stats_polar, color='green', alpha=0.25)
ax.plot(angles_polar, stats_polar, color='blue', linewidth=2)

ax.set_xticks(angles_polar[:-1])
ax.set_xticklabels(df_7_polar_area['Year'])

plt.title('Hydroelectric Energy Production Percentages Over Different Years (Polar Area Chart)', size=16, color='black', y=1.1)

plt.show()
```
