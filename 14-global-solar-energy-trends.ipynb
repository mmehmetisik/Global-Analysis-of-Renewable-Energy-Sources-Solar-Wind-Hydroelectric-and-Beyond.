# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Global Solar Energy Usage Over Years</div></b>

```python
df_14_time_series = df_14.groupby('Year')['Solar (% equivalent primary energy)'].mean().reset_index()

frames = [go.Frame(
    data=[go.Scatter(x=df_14_time_series['Year'][:k+1], y=df_14_time_series['Solar (% equivalent primary energy)'][:k+1], mode='lines+markers')],
    layout=go.Layout(title_text=f'Global Solar Energy Usage in {year}')
) for k, year in enumerate(df_14_time_series['Year'])]

initial_frame = go.Frame(
    data=[go.Scatter(x=[df_14_time_series['Year'].iloc[0]], y=[df_14_time_series['Solar (% equivalent primary energy)'].iloc[0]], mode='markers')],
    layout=go.Layout(title_text='Global Solar Energy Usage at Start')
)
frames.insert(0, initial_frame)

fig = go.Figure(
    data=[go.Scatter(x=df_14_time_series['Year'], y=df_14_time_series['Solar (% equivalent primary energy)'], mode='lines+markers')],
    layout=go.Layout(
        xaxis=dict(range=[df_14_time_series['Year'].min(), df_14_time_series['Year'].max()], autorange=False),
        yaxis=dict(range=[df_14_time_series['Solar (% equivalent primary energy)'].min(), df_14_time_series['Solar (% equivalent primary energy)'].max()], autorange=False),
        title="Global Solar Energy Usage Over Years",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args=[None, {"frame": {"duration": 500, "redraw": True}, "fromcurrent": True, "transition": {"duration": 300}}])])]
    ),
    frames=frames
)

fig.show()
```

```python
specified_countries = ['Austria', 'Belgium', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Italy', 
                      'Netherlands', 'Norway', 'Poland', 'Portugal', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom']

specified_countries_solar_data = df_14[df_14['Entity'].isin(specified_countries)]

total_solar_production_specified_countries = specified_countries_solar_data.groupby('Entity')['Solar (% equivalent primary energy)'].mean().sort_values(ascending=False)

plt.figure(figsize=(20, 8))
bars = plt.bar(total_solar_production_specified_countries.index, total_solar_production_specified_countries.values, color='grey')
plt.title('Average Solar Energy Usage of Specified Countries', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Average Solar Energy Usage (%)', fontsize=12, fontweight='bold')
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

for bar in bars:
    if bar.get_x() + bar.get_width() / 2 == total_solar_production_specified_countries.index.get_loc("Turkey"):
        bar.set_color('blue') 
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}%', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Solar Energy Production of Top Countries in Ten-Year Periods</div></b>

```python
periods = [(2000, 2009), (2010, 2020)]

production_by_period = {}

for start, end in periods:
    mask = (df_14['Year'] >= start) & (df_14['Year'] <= end)
    total_production = df_14[mask].groupby('Entity')['Solar (% equivalent primary energy)'].sum()
    production_by_period[f"{start}-{end}"] = total_production

production_df = pd.DataFrame(production_by_period)

production_df.fillna(0, inplace=True)

top_countries = production_df.sum(axis=1).sort_values(ascending=False).head(20).index
production_df_top = production_df.loc[top_countries]

production_df_top.plot(kind='bar', figsize=(15, 8))
plt.title("Solar Energy Production of Top Countries in Ten-Year Periods", fontweight='bold')
plt.xlabel("Countries", fontweight='bold')
plt.ylabel("Total Solar Energy Production (%)", fontweight='bold')
plt.legend(title="Period")
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Solar Energy Usage Percentages Over Different Years (Polar Area Chart)</div></b>

```python
df_14_polar_area = df_14.groupby('Year')['Solar (% equivalent primary energy)'].mean().reset_index()

df_14_polar_area['Year'] = df_14_polar_area['Year'].astype('category')

num_vars_polar = df_14_polar_area['Year'].nunique()

angles_polar = np.linspace(0, 2 * np.pi, num_vars_polar, endpoint=False).tolist()

stats_polar = df_14_polar_area['Solar (% equivalent primary energy)'].tolist()

stats_polar += stats_polar[:1]
angles_polar += angles_polar[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.fill(angles_polar, stats_polar, color='orange', alpha=0.25)
ax.plot(angles_polar, stats_polar, color='red', linewidth=2)

ax.set_xticks(angles_polar[:-1])
ax.set_xticklabels(df_14_polar_area['Year'])

plt.title('Solar Energy Usage Percentages Over Different Years (Polar Area Chart)', size=16, color='black', y=1.1)

plt.show()
```
