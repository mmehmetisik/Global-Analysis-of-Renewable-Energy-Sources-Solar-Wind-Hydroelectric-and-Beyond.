```python
def print_section_title(title):
    print(colored(title, 'blue', attrs=['bold', 'underline']))

def display_head_and_tail(dataframe, head=5):
    display(dataframe.head(head).style.set_caption("Head"))
    display(dataframe.tail(head).style.set_caption("Tail"))

def display_na(dataframe):
    na_df = dataframe.isnull().sum().reset_index()
    na_df.columns = ['Column', 'Number of NA']
    display(na_df.style.set_caption("Number of NA Values"))

def display_quantiles(dataframe):
    quantiles_df = dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T
    display(quantiles_df.style.format("{:.2f}").set_caption("Quantiles"))

def check_df(dataframe, head=5):
    print_section_title('Shape')
    print(dataframe.shape)
    print_section_title('Types')
    print(dataframe.dtypes.to_frame('Data Type').style.set_caption("Data Types"))
    print_section_title('Head & Tail')
    display_head_and_tail(dataframe, head)
    print_section_title('NA Values')
    display_na(dataframe)
    print_section_title('Quantiles')
    display_quantiles(dataframe)

check_df(df_4)
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Cluster Analysis of Countries Based on Renewable Energy Usage</div></b>

```python
df_cluster = df_4.groupby('Entity')['Renewables (% electricity)'].mean().reset_index()

kmeans = KMeans(n_clusters=5, n_init=10, random_state=0)

df_cluster['Cluster'] = kmeans.fit_predict(df_cluster[['Renewables (% electricity)']])

fig = px.scatter(df_cluster, x='Entity', y='Renewables (% electricity)', color='Cluster',
                 title='Cluster Analysis of Countries Based on Renewable Energy Usage',
                 labels={'Renewables (% electricity)': 'Average Renewable Energy Usage (%)'},
                 hover_data=['Entity'])

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Countries with the Highest and Lowest Renewable Energy Usage Rates (2022)</div></b>

```python
df_4_data = {
    'Entity': ['Czechia', 'Malta', 'Cyprus', 'Bulgaria', 'Poland', 'Turkey', 'Luxembourg', 'Denmark', 'Austria', 'Lithuania', 'Latvia'],
    'Year': [2022] * 11,
    'Renewables (% electricity)': [12.34, 12.44, 16.7, 18.7, 20.98, 60.39, 85.71, 83.96, 76.29, 74.82, 72.13]
}
df_4 = pd.DataFrame(df_4_data)

latest_year = df_4['Year'].max()
turkey_data = df_4[df_4['Entity'] == 'Turkey']

top_5_highest = df_4.iloc[-5:]
top_5_lowest = df_4.iloc[:5]

combined_data = pd.concat([top_5_lowest, turkey_data, top_5_highest])

colors = ['red' if entity in top_5_lowest['Entity'].values 
          else ('green' if entity in turkey_data['Entity'].values else 'blue') 
          for entity in combined_data['Entity']]

plt.figure(figsize=(20, 8))
bars = plt.bar(combined_data['Entity'], combined_data['Renewables (% electricity)'], color=colors)

for bar in bars:
    height = bar.get_height()
    if bar.get_x() + bar.get_width()/2 - 0.15 < 0:
        text_x = bar.get_x() + bar.get_width()/2
    else:
        text_x = bar.get_x() + bar.get_width()/2 - 0.15
    plt.text(text_x, height + 0.5, f'{height:.2f}', ha='center', fontsize=12, fontweight='bold')

plt.title(f'Countries with the Highest and Lowest Renewable Energy Usage Rates ({latest_year})', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=14, fontweight='bold')
plt.ylabel('Renewable Energy Rate (%)', fontsize=14, fontweight='bold')

plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tick_params(axis='both', which='major', labelsize=12, width=2)

plt.tight_layout()
plt.show()
```

    <strong>Graph Analysis:</strong> This bar chart compares the renewable energy usage rates of countries with the highest and lowest percentages, along with Turkey, for the year {latest_year}. The chart displays each country on the x-axis and their renewable energy rates on the y-axis. It provides a stark contrast between the top and bottom performers in renewable energy adoption. Including Turkey's data offers a benchmark for comparing its renewable energy efforts against these countries. The chart's design, with clear labels and a 45-degree rotation for the x-axis ticks, ensures readability.
