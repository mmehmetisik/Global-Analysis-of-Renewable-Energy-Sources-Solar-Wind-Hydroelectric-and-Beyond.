```python
def print_section_title(title):
    print(colored(title, 'blue', attrs=['bold', 'underline']))

def display_head_and_tail(dataframe, head=5):
    display(dataframe.head(head).style.set_caption("Head"))
    display(dataframe.tail(head).style.set_caption("Tail"))

def display_na(dataframe):
    na_df = dataframe.isnull().sum().reset_index()
    na_df.columns = ['Column', 'Number of NA']
    display(na_df.style.set_caption("Number of NA Values"))

def display_quantiles(dataframe):
    quantiles_df = dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T
    display(quantiles_df.style.format("{:.2f}").set_caption("Quantiles"))

def check_df(dataframe, head=5):
    print_section_title('Shape')
    print(dataframe.shape)
    print_section_title('Types')
    print(dataframe.dtypes.to_frame('Data Type').style.set_caption("Data Types"))
    print_section_title('Head & Tail')
    display_head_and_tail(dataframe, head)
    print_section_title('NA Values')
    display_na(dataframe)
    print_section_title('Quantiles')
    display_quantiles(dataframe)

check_df(df_10)
```

<a id='GlobalWindEnergyUsageOverYears'></a>
# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Global Wind Energy Usage Over Years</div></b>

```python
import plotly.graph_objects as go

df_10_time_series = df_10.groupby('Year')['Wind (% equivalent primary energy)'].mean().reset_index()

frames = [go.Frame(
    data=[go.Scatter(x=df_10_time_series['Year'][:k+1], y=df_10_time_series['Wind (% equivalent primary energy)'][:k+1], mode='lines+markers')],
    layout=go.Layout(title_text=f'Global Wind Energy Usage in {year}')
) for k, year in enumerate(df_10_time_series['Year'])]

initial_frame = go.Frame(
    data=[go.Scatter(x=[df_10_time_series['Year'].iloc[0]], y=[df_10_time_series['Wind (% equivalent primary energy)'].iloc[0]], mode='markers')],
    layout=go.Layout(title_text='Global Wind Energy Usage at Start')
)
frames.insert(0, initial_frame)

fig = go.Figure(
    data=[go.Scatter(x=df_10_time_series['Year'], y=df_10_time_series['Wind (% equivalent primary energy)'], mode='lines+markers')],
    layout=go.Layout(
        xaxis=dict(range=[df_10_time_series['Year'].min(), df_10_time_series['Year'].max()], autorange=False),
        yaxis=dict(range=[df_10_time_series['Wind (% equivalent primary energy)'].min(), df_10_time_series['Wind (% equivalent primary energy)'].max()], autorange=False),
        title="Global Wind Energy Usage Over Years",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args=[None, {"frame": {"duration": 500, "redraw": True}, "fromcurrent": True, "transition": {"duration": 300}}])])]
    ),
    frames=frames
)

fig.show()
```

```python
specified_countries = ['Austria', 'Belgium', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Italy', 
                      'Netherlands', 'Norway', 'Poland', 'Portugal', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom']

specified_countries_wind_data = df_10[df_10['Entity'].isin(specified_countries)]

total_wind_production_specified_countries = specified_countries_wind_data.groupby('Entity')['Wind (% equivalent primary energy)'].mean().sort_values(ascending=False)

plt.figure(figsize=(20, 8))
bars = plt.bar(total_wind_production_specified_countries.index, total_wind_production_specified_countries.values, color='grey')
plt.title('Average Wind Energy Usage of Specified Countries', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Average Wind Energy Usage (%)', fontsize=12, fontweight='bold')
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

for bar in bars:
    if bar.get_x() + bar.get_width() / 2 == total_wind_production_specified_countries.index.get_loc("Turkey"):
        bar.set_color('blue')
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Wind Energy Production of Top Countries in Ten-Year Periods</div></b>

```python
periods = [(2000, 2009), (2010, 2020)]

production_by_period = {}

for start, end in periods:
    mask = (df_10['Year'] >= start) & (df_10['Year'] <= end)
    total_production = df_10[mask].groupby('Entity')['Wind (% equivalent primary energy)'].sum()
    production_by_period[f"{start}-{end}"] = total_production

production_df = pd.DataFrame(production_by_period)

production_df.fillna(0, inplace=True)

top_countries = production_df.sum(axis=1).sort_values(ascending=False).head(20).index
production_df_top = production_df.loc[top_countries]

production_df_top.plot(kind='bar', figsize=(20, 10))
plt.title("Wind Energy Production of Top Countries in Ten-Year Periods", fontweight='bold')
plt.xlabel("Countries", fontweight='bold')
plt.ylabel("Total Wind Energy Production (%)", fontweight='bold')
plt.legend(title="Period")
plt.show()
```

# <b><div style='padding:15px;background-color:#004080;color:white;border-radius:15px;font-size:100%;text-align: center'>Wind Energy Usage Percentages Over Different Years (Polar Area Chart)</div></b>

```python
df_10_polar_area = df_10.groupby('Year')['Wind (% equivalent primary energy)'].mean().reset_index()

df_10_polar_area['Year'] = df_10_polar_area['Year'].astype('category')

num_vars_polar = df_10_polar_area['Year'].nunique()

angles_polar = np.linspace(0, 2 * np.pi, num_vars_polar, endpoint=False).tolist()
stats_polar = df_10_polar_area['Wind (% equivalent primary energy)'].tolist()

stats_polar += stats_polar[:1]
angles_polar += angles_polar[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.fill(angles_polar, stats_polar, color='green', alpha=0.25)
ax.plot(angles_polar, stats_polar, color='blue', linewidth=2)

ax.set_xticks(angles_polar[:-1])
ax.set_xticklabels(df_10_polar_area['Year'])

plt.title('Wind Energy Usage Percentages Over Different Years (Polar Area Chart)', size=16, color='black', y=1.1)

plt.show()
```

The gradation of green shades within the chart, coupled with the defining blue outline, illustrates the ebb and flow of wind energy's contribution to the global energy mix. This visualization method provides a unique perspective on the data, emphasizing the cyclical nature of energy trends and highlighting significant shifts in wind energy adoption. The radial distribution of years around the circle allows for an intuitive comparison of wind energy usage across different time periods, encapsulating the progressive increase in reliance on wind power.
